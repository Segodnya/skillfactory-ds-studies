```python
import h2o
from sklearn.preprocessing import LabelEncoder 
from h2o.automl import H2OAutoML
h2o.init()
```

```python
import pandas as pd

df = pd.read_csv(DIR_DATA+"mushrooms.csv")

labelEncoder = LabelEncoder()
for column in df.columns:
    df[column] = labelEncoder.fit_transform(df[column])
df = h2o.H2OFrame(df)
df = df.asfactor()
```

```python
df.head()
```

```python
# Разделяем данные на Train/Test/Validation.
# Train размером 70%, test и validation по 15% 
train_df, test_df = df.split_frame(ratios=[.7])
```

```python
y = "class"
train_columns = train_df.columns
train_columns.remove(y)

aml_model = H2OAutoML(max_runtime_secs=500, seed=1)
aml_model.train(x=train_columns, y=y, training_frame=train_df)
```

```python
leaderboard = aml_model.leaderboard
leaderboard.head()
```

```python
prediction = aml_model.predict(test_df)
```

```python
prediction
```

```python
y_true = [int(elem) for elem in test_df['class'].as_data_frame(use_pandas=True, header=True).values]
y_pred = [float(elem) for elem in prediction['p0'].as_data_frame(use_pandas=True, header=True).values]

from sklearn.metrics import roc_auc_score
roc_auc_score(y_true, y_pred)
```
